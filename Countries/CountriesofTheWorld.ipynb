```python
countries=pd.read_csv("countries of the world.csv",decimal=",")
countries.head()
```

```python
countries.rename(columns={'Infant mortality (per 1000 births)': 'infantMortality'}, inplace=True)
countries.rename(columns={'GDP ($ per capita)': 'gdp'}, inplace=True)
countries.rename(columns={'Phones (per 1000)': 'phones'}, inplace=True)

countries.head()
```

```python
#Finding correlation

#Find correlation between all values at once
corr = countries.corr()
corr
```

```python
#or do it one by one
infMAgri = countries['infantMortality'].corr(countries['Agriculture'])
infMAgri
```

```python
infMBirth = countries['infantMortality'].corr(countries['Birthrate'])
infMBirth
```

```python
gdpPhone = countries['gdp'].corr(countries['phones'])
gdpPhone
```

```python
#From these data we can see that the strongest correlation is between the Infant mortality and Birthrate (0.845764) 
#Infant Mortality and Agriculture (0.709475) 
#GDP and Phones (0.834499)
```

```python
#Visualize, style background in different colours to see good and bad correlations
plt.matshow(corr)
corr.style.background_gradient(cmap='coolwarm')
```
